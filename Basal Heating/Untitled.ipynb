```python
save_path = 'C:\\Users\\super\\Desktop\\'
df = pd.read_table(save_path+'output-ISORa1e8-256.nusselt.dat',header=None,sep="\s+")
df = df.sort_values(by=[0])
df = df.rename(columns={0:'time', 1:'nut', 2:'nub'})
df = df.reset_index(drop=True)
df['mean nut'] = df.time.map(df.groupby(['time']).nut.mean())
df['mean nub'] = df.time.map(df.groupby(['time']).nub.mean())
df = df.drop_duplicates(subset=['mean nut'])
df = df.drop(columns=['nut', 'nub'])
df = df.reset_index(drop=True)
df
```

```python
plt.rcParams['figure.figsize'] = [12, 5]
plt.plot(df['time'],df['mean nut'])
plt.axhline(df['mean nut'].mean(),color='r')
print(df['mean nut'].mean())
```

```python
plt.rcParams['figure.figsize'] = [12, 5]
plt.plot(df['time'],df['mean nub'])
plt.axhline(df['mean nub'].mean(),color='r')
print(df['mean nub'].mean())
```
